```python
import pandas as pd

filtered= pd.read_csv('./trackpy/t_allframe.csv',skiprows=1,names=['frame','y','x','size','particle'])
particleDict={}

for row in filtered.itertuples():
    if row.particle not in particleDict.keys():
        particleDict[row.particle]=[(row.frame, row.y, row.x)]
    else:
        particleDict[row.particle].append((row.frame, row.y, row.x))
```

```python
import math
import statistics

speedParticle={}
particleLife={}

for key in particleDict.keys():
    defectFrame=[]
    defectYPosition=[]
    defectXPosition=[]

    
    for item in particleDict[key]:
        defectFrame.append(item[0])
        defectYPosition.append(item[1])
        defectXPosition.append(item[2])
    
    speedParticle[key]= math.sqrt((defectYPosition[-1]-defectYPosition[0])**2+(defectXPosition[-1]-defectXPosition[0])**2)
    particleLife[key]=len(defectFrame)
max_key = max(speedParticle, key=speedParticle.get)#find the fastest particle
print(max_key, speedParticle[max_key])

with open("fastest_particle_"+str(max_key)+".txt", 'w+') as filehandle:
    for item in particleDict[max_key]:
        filehandle.write(str(item[0])+","+str(item[1])+","+str(item[2])+"\n")
```

```python
min_key= min(speedParticle, key=speedParticle.get)
print(min_key)
with open("slowest_particle_"+str(min_key)+".txt", 'w+') as filehandle:
    for item in particleDict[min_key]:
        filehandle.write(str(item[0])+","+str(item[1])+","+str(item[2])+"\n")
```

```python
#code to calculate diffusion coefficient and plot Fig11 (b)

DC={}
for key in speedParticle.keys():
    
    DC[key]= speedParticle[key]/4/(2.51304*particleLife[key])
    
DCkey=[]
DCself=[]
for key in DC.keys():
    DCkey.append(key)
    DCself.append(DC[key])

data= {'particle':DCkey, 'dc':DCself}
DC_csv= pd.DataFrame.from_dict(data)
DC_csv.to_csv("DC.csv")
```
