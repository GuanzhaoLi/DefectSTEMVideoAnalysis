```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
font_axis_publish = {
        'color':  'black',
        'weight': 'bold',
        'size': 35,
        }
labelsize = 20
fignow = plt.figure(figsize=(12,8))
ax = plt.gca()
dat = pd.read_csv("output_t_unfiltered.csv")
plt.scatter(dat['dose'] , dat['Generate'], color='r',marker = 'o',s=250,label='Generation',alpha=0.65)
plt.scatter(dat['dose'], dat['Disappear'], color='b',marker = '^',s=250,label = 'Disappearance',alpha=0.65)
plt.ylabel('Defect Counts', fontdict=font_axis_publish)
plt.legend(prop={'weight': 'bold', 'size': 15})
plt.xlabel('Dose (dpa)', fontdict=font_axis_publish)

for ticklabel in (ax.get_xticklabels()):
    ticklabel.set_fontsize(labelsize)
    ticklabel.set_fontweight('bold')

for ticklabel in (ax.get_yticklabels()):
    ticklabel.set_fontsize(labelsize)
    ticklabel.set_fontweight('bold')

#plt.show()

fignow.savefig("Dongxia_Fig10.png",format = "png",dpi=300, bbox_inches='tight', pad_inches=0.2)
```
