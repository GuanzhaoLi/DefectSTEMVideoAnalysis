```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
font_axis_publish = {
        'color':  'black',
        'weight': 'bold',
        'size': 35,
        }
labelsize = 20
# fig, ax1 = plt.subplots()
fig = plt.figure(figsize=(12,8))
ax1 = plt.gca()
dat = pd.read_csv("output_t1.csv")
ax1.scatter(dat['dose'] , dat['Generate'], color='r',marker = 'o',s=250,label='Generation',alpha=0.65)
ax1.scatter(dat['dose'], dat['Disappear'], color='b',marker = '^',s=250,label = 'Disappearance',alpha=0.65)
ax1.set_yticks(np.arange(0, 20, 4))
ax1.set_ylabel('Defect Counts', fontdict=font_axis_publish)
ax1.legend(prop={'weight': 'bold', 'size': 15},loc='upper left')
ax1.set_xlabel('Dose (dpa)', fontdict=font_axis_publish)

ax2 = ax1.twinx()
dat2 = pd.read_csv("total.csv")
ax2.scatter(dat2['dose'][::25], dat2['Total'][::25], color='k',marker = '.',s=250,alpha=1)
ax2.set_yticks(np.arange(0, 140, 20))
ax2.set_ylabel('Total Counts', fontdict=font_axis_publish)
ax2.legend(prop={'weight': 'bold', 'size': 15},loc='upper center')

for ticklabel in (ax1.get_xticklabels()):
    ticklabel.set_fontsize(labelsize)
    ticklabel.set_fontweight('bold')

for ticklabel in (ax1.get_yticklabels()):
    ticklabel.set_fontsize(labelsize)
    ticklabel.set_fontweight('bold')
    
for ticklabel in (ax2.get_yticklabels()):
    ticklabel.set_fontsize(labelsize)
    ticklabel.set_fontweight('bold')

plt.show()

fig.savefig("Dongxia_Fig_t1.png",format = "png",dpi=300, bbox_inches='tight', pad_inches=0.2)
```
